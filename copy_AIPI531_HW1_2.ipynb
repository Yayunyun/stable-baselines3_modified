```python
import gym
import numpy as np 
from stable_baselines3.a2c import A2C, MlpPolicy
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.env_util import make_vec_env
```

```python
env_id = 'CartPole-v1'
env = gym.make('CartPole-v1')
model = A2C(MlpPolicy, env, verbose=0, gae_lambda = 1, tensorboard_log='./a2c_cartpole_tensorboard/')
```

```python
# Train the model and evaluate again
# Monte Carlo estimate(high variance)
model.learn(total_timesteps=10000)
mean_reward, std_reward = evaluate_policy(model, env, n_eval_episodes=100)
print(f"mean_reward:{mean_reward:.2f} +/- {std_reward:.2f}")
```

```python
# 20 steps return # 20 steps Montecarlo
model2 = A2C(MlpPolicy, env, verbose=0, gae_lambda = 0.95, tensorboard_log='./a2c_cartpole_tensorboard/')
model2.learn(total_timesteps=10000)
mean_reward, std_reward = evaluate_policy(model2, env, n_eval_episodes=100)
print(f"mean_reward:{mean_reward:.2f} +/- {std_reward:.2f}")
```

```python
# 1 step return # 1-step estimate with bootstrapping, should have lowest variance, but high bias
model3 = A2C(MlpPolicy, env, verbose=0, gae_lambda = 0, tensorboard_log='./a2c_cartpole_tensorboard/')
model3.learn(total_timesteps=10000)
mean_reward, std_reward = evaluate_policy(model3, env, n_eval_episodes=100)
print(f"mean_reward:{mean_reward:.2f} +/- {std_reward:.2f}")
```
